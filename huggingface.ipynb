# Zero-shot Classification

Zero-shot classification is a natural language processing (NLP) task where a model predicts the class of a given input without being explicitly trained on that specific class. This is achieved by leveraging pre-trained language models to generalize across unseen categories based on their semantic meaning.

- For example, you can classify a sentence into a label like "sports" or "politics" without having labeled training data for these categories.

- (classify txt that haven't being labelled)
- instance of transfer learning

- uses pre-trained language models like OpenAI's GPT or Hugging Face's BART, RoBERTa, or T5, which are trained on massive amounts of data and understand general semantics.

Key Features:

- No Training Data Required: Works with new labels without requiring  additional data or retraining.
- Semantic Understanding: Leverages the model's deep understanding of language and concepts.
- Flexible: Can handle multi-label classification (assigning multiple categories) as well as single-label classification.

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")

classifier("This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],)

# this pipeline is called zero- shot becoz u dont need to fine tune the model on ur data to use it.
# it can directly return probabilty scores for any list of labels u want.
```

```python
classifier("I have a problem with my iphone that needs to be resolved asap!",
    candidate_labels=["urgent", "not urgent", "phone", "tablet", "computer"],)
```

# Name Entity Recognition
- model has to find which parts of the input txt correspond to entities such as person , loc, or organizations

```python
from transformers import pipeline

# Initialize the NER pipeline
ner = pipeline("ner", grouped_entities=True)

# Input text
text = "Mark Zuckerberg founded Facebook in 2004. He lives in Palo Alto, California."

# Perform NER
entities = ner(text)

# Print the results
for entity in entities:
    print(f"Entity: {entity['word']}, Type: {entity['entity_group']}, Score: {entity['score']:.2f}")
```

# Translation

```python
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")
translator("Ce cours est produit par Hugging Face.")
```

```python
from transformers import pipeline

# Create a translation pipeline for English to Marathi
en_mr_translator = pipeline("translation_en_to_xx", model="Helsinki-NLP/opus-mt-en-mr")

# Translate text
translation = en_mr_translator("How old are you?")
print(translation)
```

```python
from transformers import pipeline

# Create a translation pipeline for Marathi to English
mr_en_translator = pipeline("translation_xx_to_en", model="Helsinki-NLP/opus-mt-mr-en")

# Translate text
translation = mr_en_translator("तुम्ही कसा आहात?")
print(translation)
```
